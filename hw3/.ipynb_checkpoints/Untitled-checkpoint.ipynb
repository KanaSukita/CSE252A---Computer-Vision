```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from scipy.ndimage.filters import gaussian_filter
```

```python
def rgb2gray(rgb):
    """ Convert rgb image to grayscale.
    """
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
def corner_detect(image, nCorners, smoothSTD, windowSize):
    """Detect corners on a given image.

    Args:
        image: Given a grayscale image on which to detect corners.
        nCorners: Total number of corners to be extracted.
        smoothSTD: Standard deviation of the Gaussian smoothing kernel.
        windowSize: Window size for corner detector and non maximum suppression.

    Returns:
        Detected corners (in image coordinate) in a numpy array (n*2).

    
    """
    ### Your code here:
    i_red = int((windowSize-1)/2)
    img_smooth = gaussian_filter(image,sigma=smoothSTD)
    
    m ,n = image.shape
    dx = np.array([-1, 0, 1])
    dy = dx.T
    Ix = np.zeros(image.shape)
    Iy = np.zeros(image.shape)
    for x in range(1, n-1):
        for y in range(1, m-1):
            Ix[y, x] = sum(img_smooth[y,x-1:x+2]*dx)
            Iy[y, x] = sum(img_smooth[y-1:y+2,x]*dy)
    Ixx = Ix * Ix
    Iyy = Iy * Iy
    Ixy = Ix * Iy
    
    lamda = np.zeros(image.shape)
    
    for x in range(i_red, n-i_red):
        for y in range(i_red,m-i_red):
            Cxx = sum(map(sum,Ixx[(y-i_red):(y+i_red+1),(x-i_red):(x+i_red+1)]))
            Cyy = sum(map(sum,Iyy[(y-i_red):(y+i_red+1),(x-i_red):(x+i_red+1)]))
            Cxy = sum(map(sum,Ixy[(y-i_red):(y+i_red+1),(x-i_red):(x+i_red+1)]))
            c = np.array([[Cxx, Cxy], [Cxy, Cyy]])
            a, b = np.linalg.eig(c)
            lamda[y][x] = min(a)
    corners = np.zeros([nCorners,2])
    for i in range(nCorners):
        y, x = np.unravel_index(lamda.argmax(),lamda.shape)
        corners[i] = np.array([x, y])
        lamda[y-i_red:y+i_red+1, x-i_red:x+i_red+1] = -9999999
    
    return corners
```

```python
def ssd_match(img1, img2, c1, c2, R):
    """Compute SSD given two windows.

    Args:
        img1: Image 1.
        img2: Image 2.
        c1: Center (in image coordinate) of the window in image 1.
        c2: Center (in image coordinate) of the window in image 2.
        R: R is the radius of the patch, 2 * R + 1 is the window size

    Returns:
        SSD matching score for two input windows.

    """
    
    """
    Your code here:
    """
    matching_score = 0
    for i in range(-R, R+1):
        for j in range(-R, R+1):
            matching_score += (img1[c1[1]+i][c1[0]+j] - img2[c2[1]+i][c2[0]+j])**2
    
    return matching_score
```

```python
def naive_matching(img1, img2, corners1, corners2, R, SSDth):
    """Compute SSD given two windows.

    Args:
        img1: Image 1.
        img2: Image 2.
        corners1: Corners in image 1 (nx2)
        corners2: Corners in image 2 (nx2)
        R: SSD matching radius
        SSDth: SSD matching score threshold

    Returns:
        SSD matching result a list of tuple (c1, c2), 
        c1 is the 1x2 corner location in image 1, 
        c2 is the 1x2 corner location in image 2. 

    """
    
    """
    Your code here:
    """
    matching = []
    ssd = np.zeros([corners1.shape[0], corners2.shape[0]])
    for i in range(corners1.shape[0]):
        for j in range(corners2.shape[0]):
            ssd [i][j] = ssd_match(img1, img2, corners1[i], corners2[j], R)
            
    for i in range(corners1.shape[0]):
        minimum = min(ssd[i])
        j = np.where(ssd[i] == minimum)[0][0]
        if minimum < SSDth:
            matching.append((corners1[i], corners2[j]))
            ssd[:,j] *= 10000
    return matching
```

```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt
from scipy.io import loadmat

def compute_fundamental(x1,x2):
    """    Computes the fundamental matrix from corresponding points 
        (x1,x2 3*n arrays) using the 8 point algorithm.
        Each row in the A matrix below is constructed as
        [x'*x, x'*y, x', y'*x, y'*y, y', x, y, 1] 
    """
    
    n = x1.shape[1]
    if x2.shape[1] != n:
        raise ValueError("Number of points don't match.")
    
    # build matrix for equations
    A = np.zeros((n,9))
    for i in range(n):
        A[i] = [x1[0,i]*x2[0,i], x1[0,i]*x2[1,i], x1[0,i]*x2[2,i],
                x1[1,i]*x2[0,i], x1[1,i]*x2[1,i], x1[1,i]*x2[2,i],
                x1[2,i]*x2[0,i], x1[2,i]*x2[1,i], x1[2,i]*x2[2,i] ]
            
    # compute linear least square solution
    U,S,V = np.linalg.svd(A)
    F = V[-1].reshape(3,3)
        
    # constrain F
    # make rank 2 by zeroing out last singular value
    U,S,V = np.linalg.svd(F)
    S[2] = 0
    F = np.dot(U,np.dot(np.diag(S),V))
    
    return F/F[2,2]


def fundamental_matrix(x1,x2):
    n = x1.shape[1]
    if x2.shape[1] != n:
        raise ValueError("Number of points don't match.")

    # normalize image coordinates
    x1 = x1 / x1[2]
    mean_1 = np.mean(x1[:2],axis=1)
    S1 = np.sqrt(2) / np.std(x1[:2])
    T1 = np.array([[S1,0,-S1*mean_1[0]],[0,S1,-S1*mean_1[1]],[0,0,1]])
    x1 = np.dot(T1,x1)
    
    x2 = x2 / x2[2]
    mean_2 = np.mean(x2[:2],axis=1)
    S2 = np.sqrt(2) / np.std(x2[:2])
    T2 = np.array([[S2,0,-S2*mean_2[0]],[0,S2,-S2*mean_2[1]],[0,0,1]])
    x2 = np.dot(T2,x2)

    # compute F with the normalized coordinates
    F = compute_fundamental(x1,x2)

    # reverse normalization
    F = np.dot(T1.T,np.dot(F,T2))

    return F/F[2,2]
```

```python
def plot_epipolar_lines(img1,img2, cor1, cor2):
    """Plot epipolar lines on image given image, corners

    Args:
        img1: Image 1.
        img2: Image 2.
        cor1: Corners in homogeneous image coordinate in image 1 (3xn)
        cor2: Corners in homogeneous image coordinate in image 2 (3xn)

    """
    
    """
    Your code here:
    """
    F = fundamental_matrix(cor1, cor2)
    
    
    fig = plt.figure(figsize=(8, 8))
    plt.imshow(img1) # two dino images are of different sizes, resize one before use
    for i in range(cor1.shape[1]):
        plt.scatter(cor1[0][i], cor1[1][i], s=45, edgecolors='g', facecolors='b')
    for i in range(cor2.shape[1]):
        l_para1 = np.dot(F, cor2[:,i])
        x1 = img1.shape[1]
        #x = linspace(0, im1.shape[1]-1, im1.shape[1])
        y_start1 = l_para1[2]/(-l_para1[1])
        y_end1 = (x1*l_para1[0]+l_para1[2])/(-l_para1[1])
        plt.plot([0, x1], [y_start1, y_end1], color = 'b')
        plt.axis([0,img1.shape[1],img1.shape[0],0])
    
    plt.show()
    
    fig = plt.figure(figsize=(8, 8))
    plt.imshow(img2)
    for i in range(cor2.shape[1]):
        plt.scatter(cor2[0][i], cor2[1][i], s=45, edgecolors='g', facecolors='b')
    for i in range(cor1.shape[1]):
        l_para2 = np.dot(F.T, cor1[:,i])
        x2 = img2.shape[1]
        y_start2 = l_para2[2]/(-l_para2[1])
        y_end2 = (x2*l_para2[0]+l_para2[2])/(-l_para2[1])
        plt.plot([0, x2], [y_start2, y_end2], color = 'b')
        plt.axis([0,img2.shape[1],img2.shape[0],0])
    
    plt.show()
```

```python
def display_correspondence(img1, img2, corrs):
    """Plot matching result on image pair given images and correspondences

    Args:
        img1: Image 1.
        img2: Image 2.
        corrs: Corner correspondence

    """
    
    """
    Your code here.
    You may refer to the show_matching_result function
    """
    show_matching_result(img1, img2, corrs)

def correspondence_matching_epipole(img1, img2, corners1, F, R, SSDth):
    """Find corner correspondence along epipolar line.

    Args:
        img1: Image 1.
        img2: Image 2.
        corners1: Detected corners in image 1.
        F: Fundamental matrix calculated using given ground truth corner correspondences.
        R: SSD matching window radius.
        SSDth: SSD matching threshold.
    
    
    Returns:
        Matching result to be used in display_correspondence function

    """
    """
    Your code here.
    """
    matching = []
    n = corners1.shape[0]
    cor1 = np.row_stack((corners1.T, [1]*n))
    for i in range(n):
        l_para2 = np.dot(F.T, cor1[:,i])
        x2 = img2.shape[1]
        y_max = img2.shape[0]
        minimum = 100000000
        candidate  = np.array([-1,-1])
        for j in range(R,x2-R):
            y = int((j*l_para2[0]+l_para2[2])/(-l_para2[1]))
            if y >= R and y < y_max-R:
                cor2_temp = np.array([j, y])
                ssd = ssd_match(img1, img2, corners1[i], cor2_temp, R)
                if ssd < minimum:
                    minimum = ssd
                    candidate = cor2_temp
        if minimum < SSDth:
            matching.append((corners1[i], candidate))
    return matching
```

```python
def triangulate(corsSSD, P1, P2):
    """Find corner correspondence along epipolar line.

    Args:
        corsSSD: Corner correspondence
        P1: Projection matrix 1
        P2: Projection matrix 2
    
    Returns:
        3D points from triangulation
        
    """
    """
    Your code here
    """
    x1=[corsSSD[i][0] for i in range(len(corsSSD))]
    x2=[corsSSD[i][1] for i in range(len(corsSSD))]
    points3D=np.zeros((len(corsSSD),4))
    for i in range(len(corsSSD)):
        A1=x1[i][0]*P1[2].T-P1[0].T
        A2=x1[i][1]*P1[2].T-P1[1].T
        A3=x2[i][0]*P1[2].T-P2[0].T
        A4=x2[i][1]*P1[2].T-P2[1].T
        
        A = np.stack((A1,A2,A3,A4))
        U, S, V = np.linalg.svd(A)
        X=V[-1]
        X=X/X[3]
        points3D[i]=X
    return points3D

def inDistance(cord1, cord2, Thres):
    if (cord1[0]-cord2[0])**2+(cord1[1]-cord2[1])**2 < Thres**2:
        return True
    else:
        return False

def find_outliers(points3D, P2, outlierTH, corsSSD):
    """Find and plot outliers on image 2

    Args:
        points3D: 3D point from triangulation
        P2: Projection matrix in for image 2
        outlierTH: outlier pixel distance threshold
        corsSSD: Corner correspondence
        
    """
    """
    Your code here
    """
    x2_e=np.dot(P2, points3D.T)
    x2_e[0]=x2_e[0]/x2_e[2]
    x2_e[1]=x2_e[1]/x2_e[2]
    x2_e=x2_e[0:2].T
    outlier=[]
    inlier=[]
    outlier_e=[]
    inlier_e=[]
    x2=[corsSSD[i][1] for i in range(len(corsSSD))]
    for i in range(len(x2)):
        dis_square=(x2[i][0]-x2_e[i][0])**2+(x2[i][1]-x2_e[i][1])**2
        if dis_square<outlierTH**2:
            inlier.append(x2[i])
            inlier_e.append(x2_e[i])
        else:
            outlier.append(x2[i])
            outlier_e.append(x2_e[i])
    return outlier,inlier,outlier_e,inlier_e

def plot_liers(img2, cor2, outlier, inlier):
    fig = plt.figure(figsize=(8, 8))
    plt.imshow(img2, cmap='gray')
    for i in range(len(cor2)):
        plt.scatter(cor2[i][0], cor2[i][1], s=35, edgecolors='b', facecolors='none')
    for i in range(len(inlier)):
        plt.scatter(inlier[i][0], inlier[i][1], s=20, facecolors='r')
    for i in range(len(outlier)):
        plt.scatter(outlier[i][0], outlier[i][1], s=20, facecolors='y')
    plt.show()
```

```python
def triangulate(corsSSD, P1, P2):
    """Find corner correspondence along epipolar line.

    Args:
        corsSSD: Corner correspondence
        P1: Projection matrix 1
        P2: Projection matrix 2
    
    Returns:
        3D points from triangulation
        
    """
    """
    Your code here
    """
    n = len(corsSSD)
    points3D = np.zeros([n,4])
    for i in range(n):
        cor1 = corsSSD[i][0]
        cor2 = corsSSD[i][1]
        A1=cor1[0]*P1[2].T-P1[0].T
        A2=cor1[1]*P1[2].T-P1[1].T
        A3=cor2[0]*P2[2].T-P2[0].T
        A4=cor2[1]*P2[2].T-P2[1].T
        
        A = np.stack((A1,A2,A3,A4))
        U, S, V = np.linalg.svd(A)
        points3D[i] = V[-1]/V[-1,3]
    return points3D.T

def inDistance(cord1, cord2, Thres):
    if (cord1[0]-cord2[0])**2+(cord1[1]-cord2[1])**2 < Thres**2:
        return True
    else:
        return False

def find_outliers(points3D, P2, outlierTH, corsSSD):
    """Find and plot outliers on image 2

    Args:
        points3D: 3D point from triangulation
        P2: Projection matrix in for image 2
        outlierTH: outlier pixel distance threshold
        corsSSD: Corner correspondence
        
    """
    """
    Your code here
    """
    n = len(corsSSD)
    cor2 = [corsSSD[i][1] for i in range(n)]
    cor2_temp = np.dot(P2, points3D).astype('int')
    cor2_pred = np.zeros([n,2])
    
    inliers = []
    outliers = []
    
    for i in range(n):
        cor2_pred[i, 0] = int(cor2_temp[0, i]/cor2_temp[2, i])
        cor2_pred[i, 1] = int(cor2_temp[1, i]/cor2_temp[2, i])
        if inDistance(cor2[i], cor2_pred[i], outlierTH):
            inliers.append(cor2_pred[i])
        else:
            outliers.append(cor2_pred[i])
    return cor2, inliers, outliers

def plot_liers(img2, cor2, outlier, inlier):
    fig = plt.figure(figsize=(8, 8))
    plt.imshow(img2, cmap='gray')
    for i in range(len(cor2)):
        plt.scatter(cor2[i][0], cor2[i][1], s=35, edgecolors='black', facecolors='none')
    for i in range(len(inlier)):
        plt.scatter(inlier[i][0], inlier[i][1], s=20, facecolors='blue')
    for i in range(len(outlier)):
        plt.scatter(outlier[i][0], outlier[i][1], s=20, facecolors='red')
    plt.show()
```

```python
def corner_detect(image, nCorners, smoothSTD, windowSize):
    """Detect corners on a given image.

    Args:
        image: Given a grayscale image on which to detect corners.
        nCorners: Total number of corners to be extracted.
        smoothSTD: Standard deviation of the Gaussian smoothing kernel.
        windowSize: Window size for corner detector and non maximum suppression.

    Returns:
        Detected corners (in image coordinate) in a numpy array (n*2).

    
    """
    ### Your code here:
    i_red = int((windowSize-1)/2)
    img_smooth = gaussian_filter(image,sigma=smoothSTD)
    
    m ,n = image.shape
    dx = np.array([-1, 0, 1])
    dy = dx.T
    Ix = np.zeros(image.shape)
    Iy = np.zeros(image.shape)
    for x in range(1, n-1):
        for y in range(m-1):
            Ix[y, x] = sum(img_smooth[y,x-1:x+2]*dx)
            Iy[y, x] = sum(img_smooth[y-1:y+2,x]*dy)
    Ixx = Ix * Ix
    Iyy = Iy * Iy
    Ixy = Ix * Iy
    
    lamda = np.zeros(image.shape)
    
    for x in range(i_red, n-i_red):
        for y in range(i_red,m-i_red):
            Cxx = sum(map(sum,img_dx2[(y-i_red):(y+i_red+1),(x-i_red):(x+i_red+1)]))
            Cyy = sum(map(sum,img_dy2[(y-i_red):(y+i_red+1),(x-i_red):(x+i_red+1)]))
            Cxy = sum(map(sum,img_dxy[(y-i_red):(y+i_red+1),(x-i_red):(x+i_red+1)]))
            c = np.array([[Cxx, Cxy], [Cxy, Cyy]])
            a, b = np.linalg.eig(c)
            lamda[i][j] = min(a)
    corners = np.zeros([nCorners,2])
    for i in range(nCorners):
        y, x = np.unravel_index(lamda.argmax(),lamda.shape)
        corners[i] = np.array([x, y])
        lamda[y-i_red:y+i_red+1, x-i_red:x+i_red+1] = -9999999
    
    return corners
```

```python
def triangulate(corsSSD, P1, P2):
    """Find corner correspondence along epipolar line.

    Args:
        corsSSD: Corner correspondence
        P1: Projection matrix 1
        P2: Projection matrix 2
    
    Returns:
        3D points from triangulation
        
    """
    """
    Your code here
    """
    x1=[corsSSD[i][0] for i in range(len(corsSSD))]
    x2=[corsSSD[i][1] for i in range(len(corsSSD))]
    
    x1_homo=np.concatenate((x1,np.ones((len(corsSSD),1))),axis=1)
    x2_homo=np.concatenate((x2,np.ones((len(corsSSD),1))),axis=1)
    points3D=np.zeros((len(corsSSD),4)) #create 3D homogeneous points
    for i in range(len(corsSSD)):
        A1=(x1_homo[i,0]*P1[2].T-P1[0].T)
        A2=(x1_homo[i,1]*P1[2].T-P1[1].T)
        A3=(x2_homo[i,0]*P2[2].T-P2[0].T)
        A4=(x2_homo[i,1]*P2[2].T-P2[1].T)
        
        A=np.stack((A1,A2,A3,A4))
        [U,S,V] = np.linalg.svd(A)
        X=V[-1]
        X=X/X[3]
        points3D[i]=X
        
    return points3D

def find_outliers(points3D, P2, outlierTH, corsSSD):
    """Find and plot outliers on image 2

    Args:
        points3D: 3D point from triangulation
        P2: Projection matrix in for image 2
        outlierTH: outlier pixel distance threshold
        corsSSD: Corner correspondence
        
    """
    """
    Your code here
    """
    x2_e=np.dot(P2,points3D.T)  #x2=P2*points3D
    x2_e[0]=x2_e[0]/x2_e[2]
    x2_e[1]=x2_e[1]/x2_e[2]
    x2_e=x2_e[0:2].T
    outlier=[]
    inlier=[]
    outlier_e=[]
    inlier_e=[]
    x2=[corsSSD[i][1] for i in range(len(corsSSD))]
    for i in range(len(x2)):
        dis_square=(x2[i][0]-x2_e[i][0])**2+(x2[i][1]-x2_e[i][1])**2
        if dis_square<outlierTH**2:  #it is inlier
            inlier.append(x2[i])
            inlier_e.append(x2_e[i])
        else:
            outlier.append(x2[i])
            outlier_e.append(x2_e[i])
    return outlier,inlier,outlier_e,inlier_e
```
